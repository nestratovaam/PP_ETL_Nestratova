```python
#инициализация пустого списка
data = []

#пропускает заголовки
next(csv_reader1, None)

#читаем данные из CSV-файла построчно
for row in csv_reader1:
    if row[4] in ['Sales', 'Research & Development']:
        # Преобразование строк в числа для соответствующих столбцов
        #row[2] = float(row[2])  
        row[0] = float(row[0]) 
        row[7] = float(row[7])  
        #row[5] = float(row[5])  
        row[6] = float(row[6])  
        data.append(row)

print(len(data))

print(data[:2])
```

**Загрузка данных в SQL DB**

```python
import sqlite3

#открывает подключение к файлу базы данных SQLite
conn = sqlite3.connect('session.db')
```

```python
#удаление существующей таблицы  Life
try:
    conn.execute('DROP TABLE IF EXISTS `dataset`')
except Exception as e:
    print(str(e))
```

```python
#создание новой таблицы Life
try:
    conn.execute('''
         CREATE TABLE dataset
         (ID                            INTEGER PRIMARY KEY,
         Age                            INTEGER,
         Genger                         TEXT NOT NULL,
         Department                     TEXT NOT NULL,
         EducationField                 TEXT NOT NULL,
         MonthlyIncome                  Float DEFAULT 0);''')
    print ("Table created successfully")
except Exception as e:
    print(str(e))
    print('Table Creation Failed!!!!!')
finally:
    conn.close() #закрывает подключение к базе данных
```

```python
import sqlite3

#устанавливаем соединение с базой данных
conn = sqlite3.connect('session.db')
cursor = conn.cursor()

#выполняем запрос SQL для получения списка таблиц
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

#выводим список таблиц
print("Список таблиц:")
for table in tables:
    print(table[0])

#закрываем соединение с базой данных
conn.close()
```

```python
#дополнительные преобразования
crypto_sql_data = [(row[1], row[2], row[4], row[5], row[6]) for row in data]
crypto_sql_data[:2]
```

```python
#создание нового подключения для вставки новых данных в базу данных SQL
conn = sqlite3.connect('session.db')
cur = conn.cursor()
try:
    cur.executemany("INSERT INTO dataset(Age, Genger, Department, EducationField, MonthlyIncome) VALUES (?,?,?,?,?)", crypto_sql_data)
    conn.commit()
    print('Data Inserted Successfully')
except Exception as e:
    print(str(e))
    print('Data Insertion Failed')
finally:
    conn.close()
```

```python
#чтение данных из БД, чтобы проверить их
conn = sqlite3.connect('session.db')
rows = conn.cursor().execute('Select * from dataset')

for row in rows:
    print(row)
```
